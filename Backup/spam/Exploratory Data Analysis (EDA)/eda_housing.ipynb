```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
```

```python
# Load California housing dataset
california_housing = fetch_california_housing(as_frame=True)
california_df = california_housing.frame
```

```python
# Print basic information
print("California Housing Dataset Basic Information:")
print(california_df.head())
print(california_df.info())
print(california_df.describe())
print("\n")
```

```python
# Pairplot for California Housing dataset
sns.pairplot(california_df)
plt.title("Pairplot for California Housing Dataset")
plt.show()
```

```python
# Distribution of target variable in California Housing dataset
plt.figure(figsize=(8, 6))
sns.histplot(california_df['MedHouseVal'], bins=30, kde=True)
plt.title("Distribution of Median House Value in California Housing Dataset")
plt.xlabel("Median House Value")
plt.ylabel("Frequency")
plt.show()
```
