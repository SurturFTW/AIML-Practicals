```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load Iris dataset
iris_data = load_iris()
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris_df['target'] = iris_data.target

# Print basic information
print("Iris Dataset EDA:")
print(iris_df.head())
print(iris_df.info())
print(iris_df.describe())
print("\n")

# Pairplot for Iris dataset
sns.pairplot(iris_df, hue='target')
plt.title("Pairplot for Iris Dataset")
plt.show()

# Boxplot for Iris dataset
plt.figure(figsize=(10, 6))
for i, feature in enumerate(iris_data.feature_names):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='target', y=feature, data=iris_df)
    plt.title("Boxplot of {}".format(feature))
plt.tight_layout()
plt.show()
```
