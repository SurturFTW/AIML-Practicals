```python
import numpy as np # Linear algebra
import pandas as pd # Data processing
from sklearn.preprocessing import LabelEncoder # If data isstring then

from sklearn.tree import DecisionTreeClassifier
df = pd.read_csv('/content/sample_data/glass.csv')
df
```

```python
# Extracting Features
X = df.iloc[:,:-1]
y= df.iloc[:,9]
# Making Model
dt = DecisionTreeClassifier(criterion = "entropy")
dt
DecisionTreeClassifier(criterion='entropy')
# Splitting the dataset into training, testing and predicting values:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
y_pred
```

```python
# Making Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

**Accuracy of ID3**

```python
#finding accuracy from the confusion matrix.
a = cm.shape
corrPred = 0
falsePred = 0
for row in range(a[0]):
  for c in range(a[1]):
    if row == c:
      corrPred += cm[row,c]
    else:
      falsePred += cm[row,c]
print('Correct predictions: ', corrPred)
print('False predictions', falsePred)
print ('\n\nAccuracy of the ID3 is: ', corrPred/(cm.sum()))
```
